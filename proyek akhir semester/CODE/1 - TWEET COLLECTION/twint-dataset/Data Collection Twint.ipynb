```python
df = pd.read_csv(root_path+"vaksin-twint.csv")
```

```python
df_filtered = df[df['language']=='in']
```

```python
keywords = ['vaksin','vaksinasi','sinovac','astrazeneca','sinopharm','moderna','pfizer','novavax','sputnik-v','zifivax']
drop_idx = []

for idx, row in df_filtered.iterrows():
  tweet = row['tweet'].lower()
  is_invalid = True

  for keyword in keywords:
    if keyword in tweet:
      is_invalid = False
      break
  
  if is_invalid:
    drop_idx.append(idx)
```

```python
df_filtered.drop(drop_idx, inplace=True)
df_filtered.reset_index(drop=True, inplace=True)
```

```python
df_filtered.to_csv(root_path+'vaksin-twint-filter.csv', index=False)
```
